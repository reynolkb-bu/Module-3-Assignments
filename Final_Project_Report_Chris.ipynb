### 4.1 Analytical Framework

Describe your overall approach.

Include:
- your overall framework
- use of validation curves
- choice of MAE or RMSE (as appropriate) as the primary error metric

**4.1 Answer:**  
Our overall analytical framework followed a staged supervised learning workflow designed to move from simple benchmarks to a final production style model. The group began with a fixed train/test split so that all model development could be done on the training data while preserving a truly unseen test set for final evaluation. Within the training set, we used 5 fold repeated cross validation with 5 repeats as the main validation strategy. 

Following this, the group established a baseline performance for three candidate models, added some engineered features to measure predictive accuracy, tested reduced feature subsets to see whether feature selection was beneficial, and finally fine tuned the best model configurations before selecting a final model. This framework of work allowed for comparison of models fairly at each stage while keeping the final recommendation grounded in out of sample performance.

Validation was done by systematically testing different settings and selecting the one with the lowest VE. In part 3, the group compared feature subset sizes using quick cross validation, then confirmed the best choice with full repeated CV. In part 4, we applied the same approach to hyperparameter tuning: Ridge used a grid of alpha values while random forest and HistGradientBoosting used a two stage search. This ensured model choices were based on performance rather than intuition. 

The group used MAE as the primary metric because it directly answered the business question: how far off predictions are, on average, in dollars. We chose MAE over RMSE since RMSE can be overly influenced by a small number of very expensive homes. MAE provides a more stable and interpretable measure of typical prediction error across properties.
