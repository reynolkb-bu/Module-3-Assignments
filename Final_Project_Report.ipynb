### 4.1 Analytical Framework

Describe your overall approach.

Include:
- your overall framework
- use of validation curves
- choice of MAE or RMSE (as appropriate) as the primary error metric

**4.1 Answer:**  
Our overall framework followed a structured and iterative pipeline: define baselines, engineer features, select features, tune hyperparameters, and evaluate the final model on a held-out test set.

We split our cleaned dataset 80/20 with 80% into training and 20% into testing. We made development, feature engineering, and tuning decisions using the training data only. The test set was only used at the end to evaluate how the model performed with unseen data to determine if it could generalize well.

We used RepeatedKFold with 5 folds and 5 repeats (25 total fits per evaluation) as our primary validation. By repeating the folds with different random splits, we were able to get a stable estimate of how well the model performed. For computationally expensive steps like the feature selection sweep and hyperparameter search, we used lighter configurations (5-fold x 1 repeat or 3-fold x 1 repeat) to keep runtimes manageable. The final model was then re-evaluated with the full 5x5 setup before testing on the held-out set.

We chose Mean Absolute Error (MAE) as our primary metric because it is directly interpretable in the same units as the target variable (dollars). An MAE of $135,000 means predictions are off by $135,000 on average. This makes it easy to communicate to non-technical stakeholders. MAE is also more robust to outliers than RMSE, which squares large errors and can be influenced by a small number of extreme outliers. This was important since the dataset contains properties spanning a wide range.

We tracked MAE across the entire pipeline (baseline, feature engineering, feature selection and tuning) and monitored if each step improved the model performance. This served as our validation curve to help us confirm whether each step actually improved performance.
