### 4.1 Analytical Framework

Describe your overall approach.

Include:
- your overall framework
- use of validation curves
- choice of MAE or RMSE (as appropriate) as the primary error metric

**4.1 Answer:**  

Our framework followed a structured pipeline: establish baselines, engineer features, select features, tune hyperparameters, and evaluate the final model on a held-out test set. At each stage we tracked cross-validated MAE so we could tell whether a given step actually improved performance before committing to it.

We split the cleaned dataset 80/20 with `random_state=42`. All development, feature engineering, feature selection, and tuning decisions were made using the training set only; the 20% held-out set was touched exactly once, at the end, to measure generalization. For the primary evaluator we used `RepeatedKFold(n_splits=5, n_repeats=5)` (25 fits per evaluation) because repeated folds give a more stable CV estimate than a single 5-fold pass and make small MAE differences between models easier to trust. For computationally expensive steps we used lighter configurations to keep runtimes tractable: `RepeatedKFold(5, 1)` for the feature-selection k-sweep, and `RepeatedKFold(3, 1)` for the coarse (Stage 1) hyperparameter searches. The chosen hyperparameters were then re-scored under the full 5×5 setup before anything was reported or moved to the test set.

We benchmarked three baselines under identical CV to frame what the later numbers needed to beat: Ridge Regression at CV MAE ≈ `$149,600` (`±$1,230`), Random Forest at ≈ `$136,054` (`±$1,071`), and HistGradientBoosting at ≈ `$135,366` (`±$1,142`). The gap between Ridge and the tree-based models at baseline was our first signal that non-linear interactions and high-cardinality location features mattered more than anything a regularized linear model could capture.

For hyperparameter search we used a two-stage approach that played the same role as a formal validation curve: a coarse RandomizedSearchCV over a multi-dimensional grid (e.g., `learning_rate`, `max_iter`, `max_depth`, `min_samples_leaf` for HistGradientBoosting) under light CV, followed by a full 5×5 re-evaluation of the winning configuration. Rather than plotting a 1D train vs. CV curve for a single hyperparameter, we compared CV MAE across the sampled grid and across pipeline stages (baseline, feature engineering, feature selection, and tuning) as our empirical check on whether each change helped. We did not produce formal per-hyperparameter validation curves; instead, we relied on the stage-wise CV MAE progression and on the final train vs. CV vs. test gap to diagnose overfitting. For our final model (HistGradientBoosting), the CV MAE of `$134,437` and test MAE of `$135,442` (a `$1,006` gap) indicated that the model was generalizing, not memorizing.

We chose MAE over RMSE as the primary error metric for two reasons. First, MAE is in the same units as the target and directly interpretable. An MAE of `$135,000` means predictions are off by about `$135,000` on average, which is easy to communicate to non-technical stakeholders. Second, MAE is more robust to the extreme high-value properties in this dataset; RMSE squares large errors and would have let a handful of multi-million-dollar properties dominate the metric and the model-selection decisions. We did cross-check RMSE at the final stage as a sanity measure, but all ranking and tuning decisions were driven by MAE.

### 4.2 Data Cleaning and Preprocessing

Describe how you prepared the data.

Include:
- issues in the raw data
- handling of missing values, outliers, inconsistencies
- key decisions and why
- what worked and what did not work

**4.2 Answer:**

The raw Zillow dataset started at 77,613 rows and 55 columns and had several quality issues. Twenty-five features were missing over 70% of their values, with the worst being `buildingclasstypeid` (99.98%), `finishedsquarefeet13` (99.95%), `basementsqft` (99.94%), and `storytypeid` (99.94%). Several columns had no predictive value (`parcelid` was unique per row, `assessmentyear` was a constant 2016), and `propertyzoningdesc` held 1,907 unique free-text values that were too noisy to encode directly. The target `taxvaluedollarcnt` was heavily right-skewed with extreme outliers above `$49M`, and there was redundancy across features like `finishedsquarefeet12` and `calculatedfinishedsquarefeet`.

Our approach was to first drop irrelevant or unreliable columns, then clean problematic rows, and finally impute and encode what remained. In the first pass we dropped 7 features that were either identifiers, constants, redundant with another feature, or too high-cardinality to encode: `parcelid`, `assessmentyear`, `rawcensustractandblock`, `censustractandblock`, `propertycountylandusecode`, `propertyzoningdesc`, and `regionidcounty` (redundant with `fips`). We then applied a 70% threshold that removed 25 additional features. To handle target outliers we used IQR filtering on `taxvaluedollarcnt` (`Q1 − 1.5×IQR` to `Q3 + 1.5×IQR`, roughly `−$336K` to `$1.1M`), which removed 5,182 extreme values, plus 35 rows with a null target and 64 rows missing more than 50% of their features.

For missing values we chose three strategies based on the meaning of the feature. Absence indicators like `garagecarcnt` and `garagetotalsqft` were filled with 0, since a missing value most likely means "no garage." Categorical ID features (`airconditioningtypeid`, `heatingorsystemtypeid`, `buildingqualitytypeid`, `propertylandusetypeid`, `regionidcity`, `regionidneighborhood`, `regionidzip`, `unitcnt`, `fips`) were mode-imputed. Skewed numeric features (`lotsizesquarefeet`, `finishedsquarefeet12`, `calculatedfinishedsquarefeet`, `yearbuilt`, bathroom and bedroom counts, latitude, and longitude) were median-imputed, since the median is more robust to the skew and outliers we observed than the mean.

 For encoding, we applied one-hot encoding with `drop_first=True` to six low-cardinality categorical features to avoid multicollinearity. Three high-cardinality location features (`regionidcity`, `regionidzip`, `regionidneighborhood`) had more than 100 unique values each, so we kept them as numeric codes to avoid a dummy-variable explosion. After cleaning, imputation, and encoding, the final dataset was 72,332 rows and 64 numeric features.

What worked: IQR-based outlier removal trimmed the extreme tail while only costing us about 6.7% of the rows, and the tiered imputation strategy (zero vs. mode vs. median) was more defensible than a one-size-fits-all approach. One-hot encoding of the low-cardinality categoricals integrated cleanly with both our linear and tree-based models. What did not work as well: aggressively imputing features with very high missing rates (e.g., `airconditioningtypeid` at ~68% missing, `regionidneighborhood` at ~60%) produced columns dominated by the imputed mode and added little predictive signal. Redundant pairs like `finishedsquarefeet12` and `calculatedfinishedsquarefeet`, along with multiple bathroom-count variants, survived cleaning and had to be pruned later in feature selection. Keeping the high-cardinality region IDs as raw numeric codes worked reasonably well for tree-based models but was a weakness for linear models, which treated them as ordered quantities rather than categories.
