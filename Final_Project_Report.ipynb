### 4.1 Analytical Framework

Describe your overall approach.

Include:
- your overall framework
- use of validation curves
- choice of MAE or RMSE (as appropriate) as the primary error metric

**4.1 Answer:**  

**Pipeline at a glance:**
- Split: 80/20 train/test with `random_state=42`. The test set is touched exactly once, at the end.
- Evaluate: `RepeatedKFold(5, 5)` (25 fits) on the training set for all comparisons.
- Stages: baseline → feature engineering → feature selection → hyperparameter tuning.
- Primary metric: MAE in dollars.
- Generalization check: CV MAE vs. held-out test MAE gap.

**Cross-validation setup.**
- Split the cleaned dataset 80/20 with `random_state=42`. All development, feature engineering, feature selection, and tuning used the training set only. The held-out 20% was used exactly once, at the end.
- Primary evaluator: `RepeatedKFold(n_splits=5, n_repeats=5)`, giving 25 fits per model. Repeated folds produce a more stable CV estimate than a single 5-fold pass, which made small MAE differences between models easier to trust.
- For heavier steps we used lighter CV to keep runtimes manageable. The feature-selection *k*-sweep used `RepeatedKFold(5, 1)`. The coarse (Stage 1) hyperparameter search used `RepeatedKFold(3, 1)`.
- Each winning configuration from a light-CV pass was re-scored under the full 5×5 setup before we trusted the number.

**Baselines.** Under identical CV, the three baselines were:
- Ridge Regression: CV MAE ≈ `$149,600` (`±$1,230`)
- Random Forest: CV MAE ≈ `$136,054` (`±$1,071`)
- HistGradientBoosting: CV MAE ≈ `$135,366` (`±$1,142`)

The gap between Ridge and the tree-based models was our first signal. Non-linear interactions and high-cardinality location features mattered more than a regularized linear model could capture.

**Hyperparameter search (our validation-curve substitute).**
- Stage 1: coarse `RandomizedSearchCV` over a multi-dimensional grid under light CV. For HistGradientBoosting this covered `learning_rate`, `max_iter`, `max_depth`, and `min_samples_leaf`.
- Stage 2: re-score the winning configuration under the full 5×5 setup.
- We did not produce formal per-hyperparameter validation curves. Sweeping one hyperparameter at a time while holding the others fixed would have missed the interaction effects that actually drive tree-ensemble performance, and the compute cost was prohibitive under the full 5×5 CV. A randomized multi-dimensional search compared across pipeline stages (baseline → FE → FS → tuning) gave us the same empirical check on whether each change helped.
- For overfitting diagnostics we relied on the train vs. CV vs. test gap. For the final model (HistGradientBoosting), CV MAE was `$134,437` and test MAE was `$135,442`. The `$1,006` gap was our evidence the model generalized rather than memorized.

**Why MAE, not RMSE.**
- Interpretability. MAE is in the same units as the target. An MAE of `$135,000` means predictions are off by about `$135,000` on average. That is easy to explain to a non-technical stakeholder.
- Robustness to the right tail. RMSE squares large errors. A handful of multi-million-dollar properties would dominate the metric and our model-selection decisions. MAE weights every error equally and keeps tuning aligned with the general housing market.

### 4.2 Data Cleaning and Preprocessing

Describe how you prepared the data.

Include:
- issues in the raw data
- handling of missing values, outliers, inconsistencies
- key decisions and why
- what worked and what did not work

**4.2 Answer:**

**Raw data issues.**

The raw Zillow dataset started at 77,613 rows and 55 columns.

- Twenty-five features were missing over 70% of their values. Worst offenders: `buildingclasstypeid` (99.98%), `finishedsquarefeet13` (99.95%), `basementsqft` (99.94%), `storytypeid` (99.94%).
- Several columns had no predictive value. `parcelid` was unique per row; `assessmentyear` was a constant 2016.
- `propertyzoningdesc` held 1,907 unique free-text values, too noisy to encode directly.
- The target `taxvaluedollarcnt` was heavily right-skewed with extreme outliers above `$49M`.
- Redundancy existed across near-duplicate features like `finishedsquarefeet12` and `calculatedfinishedsquarefeet`.

**Cleaning pipeline.**

1. Drop irrelevant or unreliable columns. 7 features were dropped in the first pass as identifiers, constants, redundant with another feature, or too high-cardinality to encode: `parcelid`, `assessmentyear`, `rawcensustractandblock`, `censustractandblock`, `propertycountylandusecode`, `propertyzoningdesc`, `regionidcounty` (redundant with `fips`).
2. Drop high-null features. A 70% missing-value threshold removed 25 additional features.
3. IQR-filter the target. `Q1 − 1.5×IQR` to `Q3 + 1.5×IQR` (roughly `−$336K` to `$1.1M`) removed 5,182 extreme values.
4. Drop problematic rows. 35 rows with a null target and 64 rows missing more than 50% of their features were also removed.
5. Impute, then encode (see below).

Final dataset after cleaning: 72,332 rows, 64 numeric features.

**Imputation strategy (by feature meaning).**

- Fill with 0: absence indicators where missing means "none" (`garagecarcnt`, `garagetotalsqft`).
- Mode-impute: categorical IDs (`airconditioningtypeid`, `heatingorsystemtypeid`, `buildingqualitytypeid`, `propertylandusetypeid`, `regionidcity`, `regionidneighborhood`, `regionidzip`, `unitcnt`, `fips`).
- Median-impute: skewed numeric features (`lotsizesquarefeet`, `finishedsquarefeet12`, `calculatedfinishedsquarefeet`, `yearbuilt`, bathroom/bedroom counts, latitude, longitude).

Median beats mean here because the observed skew and outliers would have pulled a mean-based fill toward unrepresentative values.

**Encoding.**

One-hot encoding with `drop_first=True` was applied to six low-cardinality categorical features to avoid multicollinearity. Three high-cardinality location features (`regionidcity`, `regionidzip`, `regionidneighborhood`) had more than 100 unique values each and were kept as numeric codes to avoid a dummy-variable explosion.

**What worked.**

- IQR-based outlier removal trimmed the extreme tail while only costing about 6.7% of the rows.
- The tiered imputation strategy (zero / mode / median) was more defensible than a one-size-fits-all approach.
- One-hot encoding of the low-cardinality categoricals integrated cleanly with both linear and tree-based models.

**What did not work as well.**

- Imputing features with very high missing rates (`airconditioningtypeid` at ~68%, `regionidneighborhood` at ~60%) produced columns dominated by the imputed mode and added little predictive signal.
- Redundant pairs (`finishedsquarefeet12` vs. `calculatedfinishedsquarefeet`, plus multiple bathroom-count variants) survived cleaning and had to be pruned later in feature selection.
- Keeping high-cardinality region IDs as numeric codes worked for tree-based models but was a weakness for linear models, which treated them as ordered quantities rather than categories.

Describe your feature engineering.

**4.3 Answer:**  

**Engineered features.**

We added three features:

- `zip_median_sqft`: median square footage of homes within each zip code. Absolute square footage alone does not tell you how a home compares to its neighborhood, and that comparison is what buyers and appraisers care about. A 2,000 sqft home in a neighborhood of 1,200 sqft homes is very different from a 2,000 sqft home in a neighborhood of 4,000 sqft homes.
- `bath_to_bed_ratio`: bathroom count divided by bedroom count. A high ratio often signals an updated or luxury layout rather than a simple size difference.
- `property_age`: `2017 − yearbuilt`. Expresses the same information as `yearbuilt` but on a scale that is easier to read as a proxy for wear and renovation.

**Other preprocessing decisions.**

- Standardized all features with `StandardScaler` after engineering, since Ridge is sensitive to feature scale.
- Did not try log transforms, polynomial terms, or binning. Tree-based models handle non-linear patterns on their own, so extra transformations would have added complexity without a clear payoff.

**Feature selection.**

We used `SelectKBest` with a different scorer for each model: `f_regression` for Ridge, tree feature importance for Random Forest, and permutation importance for HistGradientBoosting. We swept `k` values of 5, 10, 15, 20, 30, and 45.

- Ridge: best at all 45 features.
- Random Forest: best at all 45 features.
- HistGradientBoosting: best at `k=30`.

**What worked.**

- All three engineered features survived feature selection in every model. They were ranked important enough to keep even when roughly a third of the feature set was pruned.
- `zip_median_sqft` showed up as a top predictor in HistGradientBoosting, which matched our intuition that neighborhood size context is genuinely new information.

**What did not work as well.**

- The MAE gains from feature engineering alone were small: Ridge improved by about `$513`, HistGradientBoosting by about `$390`, and Random Forest actually got slightly worse by about `$48`.
- Feature selection on its own did not help Ridge, which got worse by about `$1,400`.
- The engineered features were useful but not game-changers. The bigger improvements came later from hyperparameter tuning on the tree-based models.

### 4.4 Model Selection

Describe how you selected your model.

Include:
- models you tried and why
- how you evaluated generalization
- hyperparameter tuning
- how you chose the final model

**4.4 Answer:**

**Models tried.**

Three models, one from each family:
- Ridge Regression: regularized linear baseline. Fast, interpretable, and gives us a floor to judge whether the tree-based models earn their added complexity.
- Random Forest: bagged tree ensemble. Captures the non-linear interactions between size, location, age, and quality that housing prices depend on.
- HistGradientBoosting: boosted tree ensemble. Same motivation as Random Forest, but typically more accurate and faster at similar depth.

**How we evaluated generalization.**
- Every model was scored with the same `RepeatedKFold(5, 5)` CV on the training set. That made cross-model differences real performance differences rather than luck from a single split.
- 20% of the data was held out from the start. It was touched exactly once, at the very end, to check how closely the CV MAE matched the true unseen-data MAE.

**Hyperparameter tuning.**

For the tree-based models we used a two-stage approach. Stage 1 was a quick `RandomizedSearchCV` with a light CV setup (3-fold, few combinations) to cheaply narrow the search space. Stage 2 re-scored the winning configuration under the full `RepeatedKFold(5, 5)`. This was much faster than running a single large `GridSearchCV`. Ridge used `GridSearchCV` directly over `alpha ∈ [0.01, 0.1, 0.5, 1.0, 5.0, 10.0, 50.0, 100.0]`, since its parameter space is small.

**Final tuned results.**
- Ridge: `alpha=0.01`. CV MAE `$150,518`.
- Random Forest: `n_estimators=100, max_depth=20, max_features='sqrt', min_samples_split=10`. CV MAE `$134,856`.
- HistGradientBoosting: `learning_rate=0.1, max_iter=200, max_depth=5, min_samples_leaf=20`. CV MAE `$134,437`.

Ridge got a little worse after feature selection. That makes sense: Ridge shrinks weak features rather than zeroing them out, so removing those features took away small signals it was still using, rather than removing pure noise. For the tree-based models, both feature selection and tuning helped.

**Why HistGradientBoosting was the final choice.**
- Lowest CV MAE of the three.
- Trained several times faster than Random Forest at every stage.
- Performance was stable across folds.

We re-fit HistGradientBoosting on the full 80% training set and evaluated once on the held-out 20%. Test MAE was `$135,442`, only `$1,006` above the CV estimate, which told us the model generalizes rather than overfits to the validation folds. We accepted the loss of coefficient-level interpretability in exchange for roughly `$16,000` better MAE than Ridge, since the business goal is accurate prices rather than explainable pricing formulas.

- `$1M` home: about 13.5% error.

**5.3 Answer:**  

Our error analysis works at the aggregate level (overall MAE, fold-to-fold variance, CV-vs-test gap) rather than the residual level. We did not produce residual-vs-predicted plots or signed-bias summaries by price bucket; that limitation is listed explicitly below. The patterns described here are inferred from the aggregate metrics we did compute, which is sufficient to identify where the model is structurally weak even if it is not as precise as a direct residual analysis.

**Fold-level instability under OLS.**

During the baseline stage we ran OLS (no regularization) under `RepeatedKFold(5, 5)`. 24 of 25 folds produced normal MAE values near `$149K`. One fold blew up to an effectively nonsensical value, on the order of `$33` trillion. This was the most surprising observation in the project. OLS has no mechanism to shrink redundant coefficients, so a single unlucky fold with the wrong conditioning produced wildly unstable predictions. Switching to Ridge Regression (`alpha=0.01`) eliminated the instability entirely and gave a stable CV MAE of `$149,600` with a fold standard deviation of only `$1,230`. The lesson: redundant features and linear models do not coexist well. That finding drove our decision to rely on tree-based models and treat Ridge only as a baseline.

**Structural under-fit of Ridge.**

Ridge's tuned CV MAE of `$150,518` is roughly `$16,000` worse than HistGradientBoosting's. The gap is consistent across every stage of the pipeline (baseline, feature engineering, feature selection, tuning). That is a clear pattern of under-fitting, not over-fitting. Ridge is not failing because it memorizes the training set. It is failing because housing price depends on non-linear interactions. The effect of square footage on price is not constant across the price range. Location effects are categorical, not ordered. A linear-in-features model cannot represent either. The tree-based models closed most of that gap.

**Generalization of the final HGB model.**

For HistGradientBoosting, CV MAE was `$134,437` and test MAE was `$135,442`. The gap is `$1,006`, about 0.7% of the error itself. Across the 25 CV folds the standard deviation of MAE was `$1,194`. The single-point test MAE falls comfortably within the fold-to-fold noise of the validation procedure. There is no evidence of systematic over- or under-prediction at the aggregate level. There is no sign the test split was easier or harder than the training folds.

**Out-of-distribution risk above ~$1.1M.**

During cleaning we applied IQR filtering on `taxvaluedollarcnt`. That capped the target range at roughly `−$336K` to `$1,112,294` and removed 5,182 extreme values. Most of those were multi-million-dollar luxury homes. The raw data contained properties assessed up to about `$50M`. We made this choice deliberately to keep the model focused on the general housing market rather than being dragged by the long right tail. It also means our reported MAE does not speak to how the model performs on homes above roughly `$1.1M`. Any use of the model on luxury properties would be an extrapolation outside its training distribution. Tree-based models regress toward the bulk of the training distribution, so we would expect systematic under-prediction at the high end.

**Limitation: no formal residual analysis.**

We did not produce residual-vs-predicted plots, signed-bias-by-price-bucket summaries, or a deep look at the 10 to 20 worst predictions. The claims above about where the model breaks down are inferred from aggregate metrics, not directly measured.

**What is worth improving.**

The two most useful follow-ups are a proper residual-level analysis and a re-training experiment on `log(taxvaluedollarcnt)`. The log target was flagged during Milestone 1 EDA as a natural way to reframe the problem as percentage error rather than absolute dollar error. It would make error sizes more consistent across the price range and also change how the reported MAE is interpreted. Adding features that describe the high end of the market (comparable sales, school ratings, walkability) would further close the gap on above-median properties.

The top drivers were square footage (`calculatedfinishedsquarefeet`, `finishedsquarefeet12`), geographic location (latitude, longitude, region IDs for zip, city, and neighborhood), and property age. Bathroom and bedroom counts were secondary. Our engineered `zip_median_sqft` feature survived feature selection in all three models and ranked among HistGradientBoosting's top predictors, confirming that neighborhood-level context adds signal beyond raw square footage.
- Build and compare multiple regression models. Tree-based ensembles substantially outperformed the regularized linear baseline. HistGradientBoosting and Random Forest landed within `$420` of each other at CV MAE near `$134,500`. Ridge was about `$16,000` worse at `$150,518`.
- Evaluate whether the chosen model generalizes. HistGradientBoosting achieved a test MAE of `$135,442` against a CV MAE of `$134,437`. The `$1,006` gap (about 0.7%) is strong evidence the model is not overfitting the validation folds.

**Business and practical implications.**

- An MAE of roughly `$135,000` makes the model a first-pass valuation tool, not a replacement for a full appraisal.
- Listings whose asking price falls well outside the model's prediction can be flagged for closer review by a human agent. That triage is expensive to do manually at scale.
- Accuracy is tier-dependent. On mid-to-high-value homes, `$135K` is 10 to 15% of the price. On low-value homes, the same error can exceed 50%. The model should be deployed with a price-tier gate.
- HistGradientBoosting trained in under 10 seconds and scores new properties in near real time. It can be embedded in listing ingestion pipelines rather than run only as a batch job.

**Limitations and constraints.**

- The model was trained and evaluated only on properties within an IQR-filtered range of roughly `−$336K` to `$1,112,294`. About 5,182 extreme values, mostly multi-million-dollar luxury properties, were removed during cleaning. The model should not be used to price luxury homes without retraining.
- The dataset is limited to the physical and location attributes Zillow provided in 2017. It does not include recent comparable sales, mortgage data, school district ratings, walkability scores, or local economic indicators.
- We did not perform residual-level error analysis. Our claims about where the model breaks down are inferred from aggregate metrics, not directly measured.
- Tax-assessed value is not the same as market sale price. The two can diverge meaningfully in hot or cold markets.
- Our cross-validation did not account for temporal structure. The data is a single 2016 to 2017 snapshot, so predictions should not be assumed to stay calibrated over time without refresh.

**Future directions.**

- Re-train on a log-transformed target (`log(taxvaluedollarcnt)`), which was flagged during Milestone 1 EDA. A log target would make error sizes more consistent across the price range. It treats a `$50K` miss on a `$500K` home the same as a `$100K` miss on a `$1M` home, which is closer to how appraisers actually think about accuracy.
- Tiered modeling. Separate low-, mid-, and high-value models would let each specialize and make errors more uniform across tiers.
- Add external features. Comparable sales, school ratings, walkability scores, and local inventory/DOM statistics directly attack what the current feature set is missing, particularly at the high end.
- Run a proper residual analysis to clarify whether the remaining error is concentrated in specific price bands, zip codes, or property types.
- Further hyperparameter tuning likely has limited headroom. The bigger wins are in data, not in the model.

**Final recommendation.**

We recommend deploying the HistGradientBoosting model as an automated first-pass valuation tool for properties in the general market (roughly `$100K` to `$1.1M`). Its predicted price should serve as a reference for flagging listings whose asking price deviates significantly from the estimate. The model should not be used unmodified for luxury properties or for markets outside the training distribution. Its predictions should be periodically recalibrated against recent sales to account for market drift. The next priority for raising accuracy is adding comparable-sales and neighborhood-quality features, and retraining on a log-transformed target. Those two changes are likely to produce a larger improvement than any further tuning of the current model.
