### 4.1 Analytical Framework

Describe your overall approach.

Include:
- your overall framework
- use of validation curves
- choice of MAE or RMSE (as appropriate) as the primary error metric

**4.1 Answer:**  

**Pipeline at a glance:**
- Split: 80/20 train/test with `random_state=42`. The test set is touched exactly once, at the end.
- Evaluate: `RepeatedKFold(5, 5)` (25 fits) on the training set for all comparisons.
- Stages: baseline → feature engineering → feature selection → hyperparameter tuning.
- Primary metric: MAE in dollars. RMSE reported only as a final sanity check.
- Generalization check: CV MAE vs. held-out test MAE gap.

Cross-validation setup. We split the cleaned dataset 80/20 with `random_state=42`. All development, feature engineering, feature selection, and tuning used the training set only. The held-out 20% was used exactly once, at the end. The primary evaluator was `RepeatedKFold(n_splits=5, n_repeats=5)`, giving 25 fits per model. Repeated folds produce a more stable CV estimate than a single 5-fold pass. That stability made small MAE differences between models easier to trust. For heavier steps we used lighter CV to keep runtimes reasonable. The feature-selection *k*-sweep used `RepeatedKFold(5, 1)`. The coarse (Stage 1) hyperparameter search used `RepeatedKFold(3, 1)`. Each winning configuration was then re-scored under the full 5×5 setup before we trusted the number.

Baselines. Under identical CV, the three baselines were:
- Ridge Regression: CV MAE ≈ `$149,600` (`±$1,230`)
- Random Forest: CV MAE ≈ `$136,054` (`±$1,071`)
- HistGradientBoosting: CV MAE ≈ `$135,366` (`±$1,142`)

The gap between Ridge and the tree-based models was our first signal. Non-linear interactions and high-cardinality location features mattered more than a regularized linear model could capture.

Hyperparameter search (our validation-curve substitute). We used a two-stage approach. Stage 1 was a coarse `RandomizedSearchCV` over a multi-dimensional grid under light CV. For HistGradientBoosting this covered `learning_rate`, `max_iter`, `max_depth`, and `min_samples_leaf`. Stage 2 re-scored the winning configuration under the full 5×5 setup. We did not produce formal per-hyperparameter validation curves. Instead, we compared CV MAE across the sampled grid and across pipeline stages (baseline → FE → FS → tuning). That was our empirical check on whether each change helped. For overfitting diagnostics we relied on the train vs. CV vs. test gap. For the final model (HistGradientBoosting), CV MAE was `$134,437` and test MAE was `$135,442`. The `$1,006` gap was our evidence the model generalized rather than memorized.

Why MAE, not RMSE.
- Interpretability. MAE is in the same units as the target. An MAE of `$135,000` means predictions are off by about `$135,000` on average. That is easy to explain to a non-technical stakeholder.
- Robustness to the right tail. RMSE squares large errors. A handful of multi-million-dollar properties would dominate the metric and our model-selection
decisions. MAE weights every error equally and keeps tuning aligned with the general housing market.

### 4.2 Data Cleaning and Preprocessing

Describe how you prepared the data.

Include:
- issues in the raw data
- handling of missing values, outliers, inconsistencies
- key decisions and why
- what worked and what did not work

**4.2 Answer:**

The raw Zillow dataset started at 77,613 rows and 55 columns and had several quality issues. Twenty-five features were missing over 70% of their values, with the worst being `buildingclasstypeid` (99.98%), `finishedsquarefeet13` (99.95%), `basementsqft` (99.94%), and `storytypeid` (99.94%). Several columns had no predictive value (`parcelid` was unique per row, `assessmentyear` was a constant 2016), and `propertyzoningdesc` held 1,907 unique free-text values that were too noisy to encode directly. The target `taxvaluedollarcnt` was heavily right-skewed with extreme outliers above `$49M`, and there was redundancy across features like `finishedsquarefeet12` and `calculatedfinishedsquarefeet`.

Our approach was to first drop irrelevant or unreliable columns, then clean problematic rows, and finally impute and encode what remained. In the first pass we dropped 7 features that were either identifiers, constants, redundant with another feature, or too high-cardinality to encode: `parcelid`, `assessmentyear`, `rawcensustractandblock`, `censustractandblock`, `propertycountylandusecode`, `propertyzoningdesc`, and `regionidcounty` (redundant with `fips`). We then applied a 70% threshold that removed 25 additional features. To handle target outliers we used IQR filtering on `taxvaluedollarcnt` (`Q1 − 1.5×IQR` to `Q3 + 1.5×IQR`, roughly `−$336K` to `$1.1M`), which removed 5,182 extreme values, plus 35 rows with a null target and 64 rows missing more than 50% of their features.

For missing values we chose three strategies based on the meaning of the feature. Absence indicators like `garagecarcnt` and `garagetotalsqft` were filled with 0, since a missing value most likely means "no garage." Categorical ID features (`airconditioningtypeid`, `heatingorsystemtypeid`, `buildingqualitytypeid`, `propertylandusetypeid`, `regionidcity`, `regionidneighborhood`, `regionidzip`, `unitcnt`, `fips`) were mode-imputed. Skewed numeric features (`lotsizesquarefeet`, `finishedsquarefeet12`, `calculatedfinishedsquarefeet`, `yearbuilt`, bathroom and bedroom counts, latitude, and longitude) were median-imputed, since the median is more robust to the skew and outliers we observed than the mean.

 For encoding, we applied one-hot encoding with `drop_first=True` to six low-cardinality categorical features to avoid multicollinearity. Three high-cardinality location features (`regionidcity`, `regionidzip`, `regionidneighborhood`) had more than 100 unique values each, so we kept them as numeric codes to avoid a dummy-variable explosion. After cleaning, imputation, and encoding, the final dataset was 72,332 rows and 64 numeric features.

What worked: IQR-based outlier removal trimmed the extreme tail while only costing us about 6.7% of the rows, and the tiered imputation strategy (zero vs. mode vs. median) was more defensible than a one-size-fits-all approach. One-hot encoding of the low-cardinality categoricals integrated cleanly with both our linear and tree-based models. What did not work as well: aggressively imputing features with very high missing rates (e.g., `airconditioningtypeid` at ~68% missing, `regionidneighborhood` at ~60%) produced columns dominated by the imputed mode and added little predictive signal. Redundant pairs like `finishedsquarefeet12` and `calculatedfinishedsquarefeet`, along with multiple bathroom-count variants, survived cleaning and had to be pruned later in feature selection. Keeping the high-cardinality region IDs as raw numeric codes worked reasonably well for tree-based models but was a weakness for linear models, which treated them as ordered quantities rather than categories.

### 4.3 Feature Engineering

Describe your feature engineering.

Include:
- transformations or new features
- why you created them
- which were useful or not useful
- what worked and what did not work

**4.3 Answer:**  

We added three engineered features. The first, `zip_median_sqft`, is the median square footage of homes within each zip code. We added it because absolute square footage alone does not tell you how a home compares to its neighborhood, and that comparison is what buyers and appraisers actually care about. A 2,000 sqft home in a neighborhood of 1,200 sqft homes is very different from a 2,000 sqft home in a neighborhood of 4,000 sqft homes. The second, `bath_to_bed_ratio`, is bathroom count divided by bedroom count. A high ratio often means an updated or luxury layout rather than a simple size difference. The third, `property_age`, is `2017 yearbuilt`. It expresses the same thing as yearbuilt but on a scale that is easier to read as a proxy for wear and renovation.

We also standardized all features using `StandardScaler` after engineering, since Ridge is sensitive to feature scale. We did not try log transforms, polynomial terms, or binning. Tree-based models handle non-linear patterns on their own, so extra transformations would have added complexity without a clear payoff.

For feature selection we used `SelectKBest` with a different scorer for each model: `f_regression` for Ridge, tree feature importance for Random Forest, and permutation importance for HistGradientBoosting. We tried `k` values of 5, 10, 15, 20, 30, and 45. Ridge and Random Forest performed best with all 45 features. HistGradientBoosting performed best at `k=30`.

What worked: all three engineered features survived feature selection in every model. That means they were ranked important enough to keep even when roughly a third of the feature set was pruned. `zip_median_sqft` showed up as a top predictor in HistGradientBoosting, which matched our intuition that neighborhood size context is genuinely new information.

What did not work: the MAE gains from feature engineering alone were small. Ridge improved by about `$513`, HistGradientBoosting by about `$390`, and Random Forest actually got slightly worse by about `$48`. Feature selection on its own also did not help Ridge, which got worse by about `$1,400`. The engineered features were useful but not game-changers. The bigger improvements came later from hyperparameter tuning on the tree-based models.

### 4.4 Model Selection

Describe how you selected your model.

Include:
- models you tried and why
- how you evaluated generalization
- hyperparameter tuning
- how you chose the final model

**4.4 Answer:**  
We tested three models, each representing a different family: Ridge Regression as a regularized linear baseline, Random Forest as a bagged tree ensemble, and HistGradientBoosting as a boosted tree ensemble. Ridge was included because it is fast, interpretable, and gives us a floor to judge whether the tree-based models are actually earning their added complexity. The two tree-based models were included because housing price depends on non-linear interactions between size, location, age, and quality which is exactly what tree ensembles are built to capture.

We evaluated generalization two ways. First, every model was scored using the same RepeatedKFold(5, 5) CV on the training set, so any differences between models are real performance differences rather than luck from a single split. Second, we held out 20% of the data from the start and only used it once, at the very end, to check how closely the CV MAE matched the true unseen-data MAE.

For hyperparameter tuning we used a two-stage approach for the tree-based models. Stage one was a quick `RandomizedSearchCV` with a lighter CV setup (3-fold, few combinations) to cheaply narrow the search space. Stage two re-scored the winning configuration under the full `RepeatedKFold(5, 5)`. This was much faster than running a single large `GridSearchCV`. Ridge used GridSearchCV directly over `alpha ∈ [0.01, 0.1, 0.5, 1.0, 5.0, 10.0, 50.0, 100.0]`, since its parameter space is small. The final choices were: Ridge `alpha=0.01`; Random Forest `n_estimators=100, max_depth=20, max_features='sqrt', min_samples_split=10`; HistGradientBoosting `learning_rate=0.1, max_iter=200, max_depth=5, min_samples_leaf=20`.

At the final tuned stage, HistGradientBoosting had the lowest CV MAE at `$134,437`, Random Forest was close behind at `$134,856`, and Ridge was well behind at `$150,518`. Ridge actually got a little worse after feature selection, which makes sense because Ridge shrinks weak features rather than zeroing them out. Removing those features took away small signals it was still using, rather than removing pure noise. For the tree-based models, both feature selection and tuning helped.

We chose HistGradientBoosting as the final model for three reasons: it had the lowest CV MAE, it trained several times faster than Random Forest at every stage, and its performance was stable across folds. We then re-fit it on the full 80% training set and evaluated once on the held-out 20%. The test MAE came in at `$135,442`, only `$1,006` above the CV estimate, which told us the model generalizes rather than overfits to the validation folds. We accepted the loss of coefficient-level interpretability in exchange for roughly `$16,000` better MAE than Ridge, since the business goal is accurate prices rather than explainable pricing formulas.

**5.1 Answer:**  

We used Mean Absolute Error (MAE) as our primary metric for the reasons described in Section 4.1: it is in the same units as the target (dollars), directly interpretable, and more robust to the extreme high-end properties in the dataset than RMSE. All model comparisons, tuning decisions, and the final evaluation are reported in MAE. Lower is better.

Comparison across models (tuned, on training data via RepeatedKFold(5, 5) cross-validation):

The two tree-based models were roughly `$16,000` better than Ridge, which confirms that the signal in housing price comes from non-linear interactions between size, location, and quality that a regularized linear model cannot capture. HistGradientBoosting edged out Random Forest by only about `$420` in CV MAE (well inside both models' fold-to-fold standard deviation), but it was also several times faster to train, which made it the clear practical choice. Validation vs test performance for the final model:
The final HistGradientBoosting model had a CV MAE of `$134,437` on the training set and a test MAE of `$135,442` on the held-out 20% it had never seen. The gap between the two is only `$1,006`, or roughly `0.7%` of the error itself. A tight gap like this is the key evidence that the model is generalizing rather than overfitting to the validation folds. If CV had said `$134K` and the test set came back at `$150K`, that would have been a red flag that our cross-validation was too optimistic or that we had leaked information during tuning. We did not see that.

What the numbers mean in practical terms:
An MAE of roughly `$135,000` means that, on average, the model's predicted price is off by about `$135,000` in either direction from the true tax-assessed value. This sounds large in isolation, but the dataset spans everything from starter homes under `$100K` to homes above `$1M`. In relative terms, `$135K` is about `27%` error on a `$500K` home, `13.5%` on a `$1M` home, and under `7%` on a `$2M` home. The model is therefore most useful as a first-pass valuation tool for mid-to-high-value properties and is not reliable for lower-end homes where the error can exceed the actual price difference between listings. This matches the recommendation given in the executive summary.

### 5.3 Error Analysis (5 points)

Include:
- patterns in residuals or prediction errors
- overprediction or underprediction
- outliers or unusual observations
- anything surprising or worth improving

**5.3 Answer:**  
Our error analysis draws on three sources in the milestones: the baseline fold-level behavior in Milestone 2, the CV-vs-test gap on the final model, and the target distribution and outlier structure established in Milestone 1.

Fold-level variability and an early warning sign. During the baseline stage, we ran OLS (no regularization) under RepeatedKFold(5, 5) and saw 24 of 25 folds produce normal MAE values near `$149K`, while a single fold blew up to an effectively nonsensical value (on the order of `$33` trillion). This was the most surprising observation in the project and was a direct consequence of multicollinearity between near-duplicate features like `calculatedfinishedsquarefeet` and `finishedsquarefeet12`. OLS has no mechanism to shrink redundant coefficients, and a single unlucky fold with the wrong conditioning produced wildly unstable predictions. Switching to Ridge Regression (`alpha=0.01`) eliminated the instability entirely and gave a stable CV MAE of `$149,600` with a fold standard deviation of only `$1,230`. The lesson was that redundant features and linear models do not coexist well, and this drove our later decision to rely on tree-based models for production and to treat Ridge only as a baseline.

Structural under-fitting of the linear model. Ridge's tuned MAE of `$150,518` is roughly `$16,000` worse than HistGradientBoosting's, and the gap is consistent across every stage of the pipeline (baseline, feature engineering, feature selection, tuning). This is a clear pattern of under-fitting rather than over-fitting: Ridge is not failing because it memorizes the training set, it is failing because housing price depends on non-linear interactions (the effect of square footage on price is not constant across the price range, and location effects are categorical rather than ordered) that a linear-in-features model cannot represent. The tree-based models closed most of that gap, which is consistent with the kind of error pattern the dataset creates.

Generalization of the final model. For the chosen HistGradientBoosting model, CV MAE was `$134,437` and test MAE was `$135,442`, a gap of only `$1,006`, or about `0.7%` of the error itself. Across the 25 CV folds the standard deviation of MAE was `$1,194`, meaning the single-point test MAE falls comfortably within the normal fold-to-fold noise of the validation procedure. There is no evidence of systematic over- or under-prediction at the aggregate level, and no sign that the test split happened to be easier or harder than the training folds.

Known limitation: the model was trained and evaluated inside a bounded price range. During cleaning, we applied IQR filtering on `taxvaluedollarcnt` that capped the target range at roughly `-$336K` to `$1,112,294` and removed 5,182 extreme values, most of them multi-million-dollar luxury homes (the raw data contained properties assessed up to about `$50M`). This decision was made deliberately to keep the model focused on the general housing market rather than being dragged by the long right tail, but it also means our reported MAE does not speak to how the model would perform on homes above roughly `$1.1M`. Any use of the model on luxury properties would be an extrapolation outside the distribution it was trained on, and would likely produce systematic under-prediction due to tree-based models' tendency to regress toward the bulk of the training distribution.

What is worth improving. The two most useful follow-ups would be a proper residual-level analysis (residuals vs. predicted, signed bias by price bucket, and examination of the 10–20 worst predictions) and a re-training experiment on `log(taxvaluedollarcnt)`, which was flagged during Milestone 1 EDA as a natural way to reframe the problem as percentage error rather than absolute dollar error. The log target would likely reduce heteroscedasticity in the residuals and produce more comparable error magnitudes across the price range, although it would also change how the reported MAE is interpreted. Adding features that specifically describe the high end of the market (comparable sales, school ratings, walkability) would further close the remaining gap for above-median properties.

On model comparison, tree-based ensembles substantially outperformed a regularized linear baseline: HistGradientBoosting and Random Forest landed within `$420` of each other at CV MAE near `$134,500`, while Ridge was about `$16,000` worse at `$150,518`. On generalization, the final HistGradientBoosting model achieved a test MAE of `$135,442` against a CV MAE of `$134,437`, a gap of only `$1,006` (about 0.7%), which is strong evidence that the model is not overfitting the validation folds.

Business and practical implications. For a real estate company, an MAE of roughly $135,000 is practically useful as a first-pass valuation tool rather than as a replacement for a full appraisal. Listings whose asking price falls well outside the model's prediction band can be flagged for closer inspection by a human agent, which is exactly the kind of triage that is expensive to do manually at scale. The model is most accurate in relative terms for mid-to-high-value homes (where `$135K` is on the order of `10–15%` of the price) and least accurate for low-value homes (where the same error can exceed 50% of the price). It should therefore be deployed with a price-tier gate rather than used uniformly across all listings. The model is also fast: HistGradientBoosting trained in under 10 seconds and can score new properties in near real time, which makes it suitable for embedding in listing ingestion pipelines rather than only running as a batch job.

Limitations and constraints. The most important limitation is that the model was trained and evaluated only on properties within an IQR-filtered range of roughly `-$336K` to `$1,112,294`. Approximately 5,182 extreme values, most of them multi-million-dollar luxury properties, were removed during cleaning. This keeps the model focused on the general housing market but means it should not be used to price luxury homes without retraining. Second, the dataset is restricted to the physical and location attributes Zillow provided in 2017. It does not include recent comparable sales, mortgage data, school district ratings, walkability scores, or local economic indicators, all of which real appraisers rely on. Third, we did not perform residual-level error analysis (residuals by price bucket, geographic error maps, signed-bias analysis), so our claims about where the model breaks down are inferred from aggregate metrics rather than directly measured. Fourth, the tax-assessed value we predicted is not the same as the market sale price, and the two can diverge meaningfully in hot or cold markets. Finally, our cross-validation did not account for temporal structure; since the data comes from a single 2016–2017 snapshot, the model has no exposure to how housing prices drift over time, and its predictions should not be assumed to stay calibrated without refresh.

Potential improvements and future directions. The most promising improvement would be re-training on a log-transformed target (`log(taxvaluedollarcnt)`), which was flagged during Milestone 1 EDA. A log target reframes the loss in percentage terms and would likely reduce the heteroscedasticity we expect in the residuals across price ranges. A second direction is tiered modeling: training separate models for low-value, mid-value, and high-value segments would let each model specialize and would make errors more uniform across tiers. Third, adding external features (recent comparable sales, school ratings, walkability scores, local inventory and DOM statistics) would directly attack what the current feature set is missing, particularly at the high end where property characteristics alone do not explain price. Fourth, a proper residual analysis would clarify whether the remaining error is concentrated in specific price bands, zip codes, or property types, which would guide the next round of feature work. Finally, given the small CV-to-test gap we already see, there is probably limited headroom from further hyperparameter tuning on the current feature set; the bigger wins are in data, not in the model.

Final recommendation. We recommend that the real estate company deploy the HistGradientBoosting model as an automated first-pass valuation tool for properties in the general market (approximately `$100K` to `$1.1M`), using the predicted price as a reference point for flagging listings whose asking price deviates significantly from the model's estimate. The model should not be used unmodified for luxury properties or for markets outside the training distribution, and its predictions should be periodically recalibrated against recent sales to account for market drift. The immediate next steps to raise accuracy further are adding comparable-sales and neighborhood-quality features to the training data, and retraining on a log-transformed target. These two changes are likely to produce a larger improvement than any further tuning of the existing model would.
